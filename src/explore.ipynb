```python
categorical_features = ['neighbourhood', 'room_type', 'neighbourhood_group', 'price']

fig, axis = plt.subplots(2, 2, figsize = (10, 7))

# Creating a multiple figure with histograms and box plots
sns.histplot(ax = axis[0, 0], data = total_data, x = "neighbourhood")
sns.histplot(ax = axis[1, 0], data = total_data, x = "room_type")
sns.histplot(ax = axis[0, 1], data = total_data, x = "neighbourhood_group")
sns.histplot(ax = axis[1, 1], data = total_data, x = "price")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

Analysis on Numeric Variables

```python
numerical_features = ['number_of_reviews', 'minimum_nights', 'reviews_per_month', 'price']

fig, axis = plt.subplots(4, 2, figsize = (10, 10), gridspec_kw={'height_ratios': [6, 1, 6, 1]})

# Creating a multiple figure with histograms and box plots
sns.histplot(ax = axis[0, 0], data = total_data, x = "price").set(xlabel = None)
sns.boxplot(ax = axis[1, 0], data = total_data, x = "price")
sns.histplot(ax = axis[0, 1], data = total_data, x = "number_of_reviews").set(xlabel = None, ylabel = None)
sns.boxplot(ax = axis[1, 1], data = total_data, x = "number_of_reviews")

sns.histplot(ax = axis[2, 0], data = total_data, x = "minimum_nights").set(xlabel = None, ylabel = None)
sns.boxplot(ax = axis[3, 0], data = total_data, x = "minimum_nights")
sns.histplot(ax = axis[2, 1], data = total_data, x = "reviews_per_month").set(xlabel = None, ylabel = None)
sns.boxplot(ax = axis[3, 1], data = total_data, x = "reviews_per_month")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
fig, axis = plt.subplots(4, 2, figsize = (10, 15))

# Create a multiple scatter diagram
sns.regplot(ax = axis[0, 0], data = total_data, x = "number_of_reviews", y = "minimum_nights")
sns.heatmap(total_data[["number_of_reviews", "minimum_nights"]].corr(), annot = True, fmt = ".2f", ax = axis[1, 0], cbar = False)

sns.regplot(ax = axis[0, 1], data = total_data, x = "SibSp", y = "minimum_nights").set(ylabel=None)
sns.heatmap(total_data[["reviews_per_month", "minimum_nights"]].corr(), annot = True, fmt = ".2f", ax = axis[1, 1])

sns.regplot(ax = axis[2, 0], data = total_data, x = "Peviews_per_month", y = "minimum_nights").set(ylabel=None)
sns.heatmap(total_data[["reviews_per_month", "minimum_nights"]].corr(), annot = True, fmt = ".2f", ax = axis[3, 0], cbar=False)

sns.regplot(ax = axis[2, 1], data = total_data, x = "price", y = "minimum_nights").set(ylabel=None)
sns.heatmap(total_data[["price", "minimum_nights"]].corr(), annot = True, fmt = ".2f", ax = axis[3, 1])



# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```
